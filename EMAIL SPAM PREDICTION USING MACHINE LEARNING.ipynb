# **EMAIL SPAM PREDICTION USING MACHINE LEARNING**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Pre-Processing

```python
data=pd.read_csv('/content/spam.csv', encoding='latin')
data
```

```python
data.isnull().sum()
```

```python
mailData = data.where((pd.notnull(data)),'')
mailData.head()
```

```python
mailData.shape
```

```python
mailData=mailData.rename(columns={'v1':'Category','v2':'Message'})
mailData.head()
```

**Label Encoding**

```python
mailData.loc[mailData['Category'] == 'spam', 'Category',] = 0
mailData.loc[mailData['Category'] == 'ham', 'Category',] = 1
```

```python
x = mailData['Message']
y = mailData['Category']
x
```

```python
y
```

**Splitting the data into training and testing**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

Feature Extraction

```python
extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase=True)
x_train_feature = extraction.fit_transform(x_train)
x_test_feature = extraction.transform(x_test)

y_train = y_train.astype('int')
y_test = y_test.astype('int')
y_train
```

```python
y_test
```

```python
print(x_train_feature)
```

Training the model

```python
model = LogisticRegression()
```

```python
model.fit(x_train_feature, y_train)
```

Evaluation

```python
predict_training_data = model.predict(x_train_feature)
accuracy_of_training_data = accuracy_score(y_train, predict_training_data)
```

```python
print('Accuracy on training the data : ', accuracy_of_training_data)
```

```python
predict_test_data = model.predict(x_test_feature)
accuracy_of_testing_data = accuracy_score(y_test, predict_test_data)
print('Accuracy on testing the data : ', accuracy_of_testing_data)
```

Building a Predictive System

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]
input_feature = extraction.transform(input_mail)

prediction = model.predict(input_feature)
prediction

if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
